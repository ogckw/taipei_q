```python
#get data from website
import requests
headers = {
'Accept':'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
'Accept-Encoding':'gzip, deflate, sdch',
'Accept-Language':'zh-TW,zh;q=0.8,en-US;q=0.6,en;q=0.4',
'Cache-Control':'max-age=0',
'Connection':'eep-alive',
'Host':'taqm.epa.gov.tw',
'Upgrade-Insecure-Requests':'1',
'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/45.0.2454.101 Safari/537.36'}
res = requests.get('http://data.taipei/opendata/datalist/datasetMeta/download?id=961ca397-4a59-45e8-b312-697f26b059dc&rid=190796c8-7c56-42e0-8068-39242b8ec927',headers = headers)


```

```python
#use the bs4
from bs4 import BeautifulSoup as bs
soup = bs(res.text)
```

```python
#Put into dic and array
mess_ary=[]
dic = {'update_date':'','update_time':'','qua_id':'','code_name':'','longitude':'','latitude':'','qua_cntu':
       '','qua_cl':'','qua_ph':''}
for s in soup.select('qua_data'):
     mess_ary.append({
            'update_date':s.select('update_date')[0].text,'update_time':s.select('update_time')[0].text,'qua_id':s.select('qua_id')[0].text,
                'code_name':s.select('code_name')[0].text,
                'longitude':s.select('longitude')[0].text,
                'latitude':s.select('latitude')[0].text,
             'qua_cntu':s.select('qua_cntu')[0].text,'qua_cl':s.select('qua_cl')[0].text,'qua_ph':s.select('qua_ph')[0].text})          
```

```python
for s in mess_ary:
    print s

#change to dataframe
import pandas as pd
waterQuality = pd.DataFrame(mess_ary)
waterQuality.index.name = 'id'
waterQuality
```

```python
#output the data
waterQuality.to_csv('waterQuality.csv',encoding='utf-8')
```
