- 載入台北市邊界資料


- https://sheethub.com/gist-map.motc.gov.tw/%E7%B8%A3%E5%B8%82%E7%95%8C/uri/75263294

```python
#先下載試跑
import json
with open('taipei.json') as f:
    data = json.load(f)
```

```python
predata = data['features'][0]['geometry']['geometries'][0]['coordinates'][0]
```

```python
#重新複習pandas 跟dictionary 其實有簡單寫法 直接write line by line and first column即可
ary = []

for pos in predata:
    dic = {'lat':'','log':''}
    dic['log'] = pos[0]
    dic['lat'] = pos[1]
    ary.append(dic)
```

```python
import pandas as pd

data = pd.DataFrame(ary)
```

```python
data.to_csv('taipei.csv',encoding='utf-8')
```

```python
# 照需求 切0.002 從左下角到右上角 (西南方到東北方)
# 用 R 跑 台北市最小緯度 24.960 最大緯度 25.210 最小經度 121.457 最大經度 121.665 第三位後無條件捨去
# 正式切割方塊 
b = 24.960 #(起始緯度)
dim = 0.002
count = 0 
for j in range(1,125):
    a = 121.457 #(起始經度)
    for i in range(1,105):
        print '{},{}'.format(a,b)  , '{},{}'.format(a+dim,b) , '{},{}'.format(a,b+dim)  , '{},{}'.format(a+dim,b+dim) , ' row = ' + str(j) , 'col = ' + str(i)  
        a = a + dim #兩百公尺
        count = count + 1
    b = b + dim
print '總共格數 = ', count
```
